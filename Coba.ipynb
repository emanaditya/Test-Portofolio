```python
#import library dulu

import pandas as pd
```

```python
#import data

df_rumah = pd.read_csv('datasets_108980_260251_train.csv')
```

```python
# cek-cek ombak dulu datanya
df_rumah.head()
```

```python
#cek kolomnya ada apa aja

df_rumah.info()
```

```python
#Kita slice dulu datanya jadi data training dan data test

df_train = df_rumah.loc[df_rumah['Id']<=999]
df_test = df_rumah.loc[df_rumah['Id']>999]

df_train.head()
```

```python
# Kita tentukan target prediksinya.

y = df_train.SalePrice
```

```python
# Kita tentukan X nya.
# Pertama kita tentukan Features-nya
```

```python
# Pilih Featuresnya

features = ["LotArea","YearBuilt", "1stFlrSF", "2ndFlrSF", "FullBath", "BedroomAbvGr", "TotRmsAbvGrd"]

# Here comes the X
X = df_train[features]
```

```python
# Cek dan ricek dulu X nya

X.head()
```

```python
# Aaannnddd... now is the time to do the modeling
# First of all, let's import the model

from sklearn.tree import DecisionTreeRegressor
mistik = DecisionTreeRegressor(random_state=10)

# Oke, sekarang model udah siap dengan nama 'mistik'
# Sekarang waktunya kita melakukan pelatihan terhadal 'mistik' dengan cara memasukan data train kita ke 'mistik'

mistik.fit(X,y)
```

```python
# Pelatihan 'mistik' sudah selesai, sekarang waktunya melakukan prediksi terhadap data test.

Prediksi = mistik.predict(df_test[features].head(10))
Prediksi
```
